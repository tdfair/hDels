# Assign hDel-v1 sgRNAs to 500-bp genomic windows

```python
hdel_sgrnas = hdel_mini.loc[hdel_mini['set'].isin(['hDels', 'flanking'])]
hdel_sgrnas.loc[:, 'adjusted_start'] = hdel_sgrnas['start'] - 250
hdel_sgrnas
```

# New hDel-v1 sgRNA bin assignments

Flanking and hDel-targeting sgRNAs with same start position have *different* bins.<br>

**For hDels without flanking sgRNAs**: hDel-targeting sgRNAs start at bin 0.<br>
**For hDels with flanking sgRNAs**: hDel-targeting sgRNAs start at bin 5.<br>

Flanking sgRNAs target human-conserved sequences flanking hDels. 

```python
#assign sgRNAs to 50-bp bins

hdel_sgrnas['50bp_bin'] = pd.cut(hdel_sgrnas['adjusted_start'], np.arange(-250, 300000, 50), include_lowest=True, right=False, labels=False)
hdel_sgrnas['50bp_bin'] = hdel_sgrnas['50bp_bin'] - 5
hdel_sgrnas['50bp_bin'].min()
```

```python
#separate hDels targeting with flanking sgRNAs

hDels_with_flank = hdel_sgrnas.loc[hdel_sgrnas['set'] == 'flanking']
hDels_no_flank = hdel_sgrnas.loc[hdel_sgrnas['set'] == 'hDels']
```

```python
#assign sgRNA to overlapping 500-bp windows (250-bp step size)

windows_flank = []

for n in range(hDels_with_flank['50bp_bin'].min(), hDels_with_flank['50bp_bin'].max(), 5):
    frame = hDels_with_flank.loc[hDels_with_flank['50bp_bin'].between(n, n+9)]
    frame['500bp_window'] = int(n*50)
    windows_flank.append(frame)
    
windows_no_flank = []

for n in range(hDels_no_flank['50bp_bin'].min(), hDels_no_flank['50bp_bin'].max(), 5):
    frame = hDels_no_flank.loc[hDels_no_flank['50bp_bin'].between(n, n+9)]
    frame['500bp_window'] = int(n*50)
    windows_no_flank.append(frame)
    
windows_flank = pd.concat(windows_flank)
windows_no_flank = pd.concat(windows_no_flank)

#unique window IDs

windows_flank['hDel_window'] = windows_flank['hDel_ID'] + '__' + windows_flank['500bp_window'].astype(str)
windows_no_flank['hDel_window'] = windows_no_flank['hDel_ID'] + '__' + windows_no_flank['500bp_window'].astype(str)
```

```python
joined_windows = pd.concat([windows_flank, windows_no_flank])
joined_windows['hDel_window'].nunique()
```

```python
window_size = joined_windows.groupby('hDel_window').size().to_frame().rename(columns={0: 'sgRNAs_per_window'})

#at least 5 sgRNAs per window

window_size_5 = window_size.loc[window_size['sgRNAs_per_window'] >= 5]

hdel_sgrnas_overlapping_5 = joined_windows.merge(window_size_5, left_on='hDel_window', right_index=True)
hdel_sgrnas_overlapping_5[['sg_ID', 'hDel_window', 'sgRNAs_per_window']]
```

```python
#aggregate sgRNAs that appear in overlapping sliding windows (500-bp, 250-bp step size) and separate by commas for Î±-RRA

hdel_windows_csv_5 = hdel_sgrnas_overlapping_5.groupby('sg_ID')['hDel_window'].agg(','.join).to_frame().reset_index()
hdel_windows_csv_5
```

# Use Î±-RRA to combine sgRNA p-values into gene FDRs

Î±-RRA is called from MAGeCK (Li et al., *Genome Biol* 2014)

```python
hdel_rra_5 = hdel_deseq.merge(hdel_windows_csv_5, on='sg_ID', how='right')

#add negative-control sgRNAs

ctrl_sgrnas = hdel_deseq.loc[hdel_deseq['sg_ID'].str.contains('panTro6_CTRL')]
ctrl_sgrnas['hDel_window'] = 'negative_control'

rra_5 = pd.concat([hdel_rra_5, ctrl_sgrnas])
rra_5
```

```python
#Î±-RRA formatting

rra_5['RRA'] = 1

#Î±-RRA input file format, sgRNAs rankings are based on adjusted p-values from DESeq2 

rra_5[['sg_ID', 'hDel_window', 'RRA', 'padj', 'RRA', 'RRA']].to_csv('~/Documents/hDels/hDel-v1_DESeq2_Î±-RRA/hDel-v1_DESeq2_Î±-RRA_5sgRNAs_input.txt', sep='\t', index=False, header=False)

#negative-control sgRNAs for Î±-RRA

rra_5.loc[rra_5['hDel_window'] == 'negative_control']['sg_ID'].to_csv('/home/tyler/Documents/hDels/hDel-v1_DESeq2_Î±-RRA/hDel-v1_negative_control_sgRNAs.txt', sep='\n', index=False, header=False)
```

```python
#p-value cutoff for setting Î±

p = 0.01

display(rra_5.loc[rra_5['padj'] < p].shape[0] / rra_5.shape[0])
```
