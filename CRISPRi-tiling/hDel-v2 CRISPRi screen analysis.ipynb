# Import hDel-v2 DESeq2 sgRNA-level analysis (design = ~time)

```python
hdel_deseq = pd.read_csv('/2TBevo/hDel-v2_C3624K_FASTQs/TFAP08-P_3lanes/hDel-v2_DESeq2.csv')
hdel_deseq.rename(columns={'Unnamed: 0': 'sg_ID', 'log2FoldChange': 'DESeq2_log2FC'}, inplace=True)
hdel_deseq
```

```python
hdel_v2 = hdel_v2.merge(hdel_deseq)
hdel_v2
```

# Filter enriched sgRNAs initially present at low abundance

```python
hdel_v2 = hdel_v2.loc[~((hdel_v2['t0_C3624K'] <= hdel_v2['t0_C3624K'].quantile(0.05)) & (hdel_v2['DESeq2_log2FC'] >= 1))]
hdel_v2
```

# Filter sgRNAs containing C4/G4 nucleotide homopolymers

```python
from re import finditer

c4_position = []
c4_sgrna = []

g4_position = []
g4_sgrna = []

for sgRNA in hdel_v2['G_N19']:
    
    for match in finditer('CCCC', sgRNA):
        c4_position.append(match.start())
        c4_sgrna.append(sgRNA)
        
    for match in finditer('GGGG', sgRNA):
        g4_position.append(match.start())
        g4_sgrna.append(sgRNA)
        
c4_sgrnas = pd.DataFrame({'G_N19': c4_sgrna, 'C_homopolymer_position': c4_position}).sort_values('C_homopolymer_position')
g4_sgrnas = pd.DataFrame({'G_N19': g4_sgrna, 'G_homopolymer_position': g4_position}).sort_values('G_homopolymer_position')

#filter polyC at positions 10-12
#filter polyG ≥ position 6

drop_sgrnas = list(set(c4_sgrnas.loc[c4_sgrnas['C_homopolymer_position'].between(10,12)]['G_N19'].to_list() + g4_sgrnas.loc[g4_sgrnas['G_homopolymer_position'] >= 6]['G_N19'].to_list()))
len(drop_sgrnas)
```

```python
hdel_v2 = hdel_v2.loc[~hdel_v2['G_N19'].isin(drop_sgrnas)]
```

# Assign hDel-v2 sgRNAs to 250-bp genomic windows

```python
hdel_sgrnas = hdel_v2.loc[(hdel_v2['set'] == 'hDels') & 
                          (hdel_v2['JostCRISPRi_specificityscore'] >= 0.20) & 
                          (hdel_v2['t0_C3624K'] >= hdel_v2['t0_C3624K'].quantile(0.01))]
```

```python
#assign sgRNA to non-overlapping 250-bp windows

windows = []

for n in range(hdel_sgrnas['50bp_bin'].astype(int).min(), hdel_sgrnas['50bp_bin'].astype(int).max(), 5):
    frame = hdel_sgrnas.loc[hdel_sgrnas['50bp_bin'].between(n, n+4)]
    frame['250bp_window'] = int(n*50)
    windows.append(frame)
        
windows_df = pd.concat(windows)

#unique window IDs

windows_df['hDel_window'] = windows_df['hDel_ID'] + '__' + windows_df['250bp_window'].astype(str)
windows_df['hDel_window'].nunique()
```

```python
windows_df = windows_df.groupby('hDel_window').filter(lambda x: len(x) >= 5)
windows_df['hDel_window'].nunique()
```

```python
hdel_windows_csv = windows_df.groupby('sg_ID')['hDel_window'].agg(','.join).to_frame().reset_index()
hdel_windows_csv
```

# Use α-RRA to combine sgRNA p-values into gene FDRs

α-RRA is called from MAGeCK (Li et al., *Genome Biol* 2014)

```python
#add hDel windows, drop sgRNAs in windows with <5 sgRNAs

hdel_sgrnas_pass = hdel_sgrnas.merge(hdel_windows_csv, how='right')

#DESeq2 negative-control sgRNAs

ctrl_sgrnas = hdel_deseq.loc[hdel_deseq['sg_ID'].str.contains('panTro6_CTRL')]
ctrl_sgrnas['hDel_window'] = 'negative_control'

hdel_sgrnas_pass = hdel_sgrnas_pass[ctrl_sgrnas.columns]
```

```python
rra = pd.concat([hdel_sgrnas_pass, ctrl_sgrnas])
rra
```

```python
#α-RRA formatting

rra['RRA'] = 1

#α-RRA input file format, sgRNAs rankings are based on adjusted p-values from DESeq2 

rra[['sg_ID', 'hDel_window', 'RRA', 'padj', 'RRA', 'RRA']].to_csv('~/Documents/hDels/hDel-v2-α-RRA/hDel-v2_DESeq2_α-RRA_in.txt', sep='\t', index=False, header=False)

#identify negative-control sgRNAs for α-RRA

rra.loc[rra['hDel_window'] == 'negative_control']['sg_ID'].to_csv('/home/tyler/Documents/hDels/hDel-v2-α-RRA/hDel-v2_negative_control_sgRNAs.txt', sep='\n', index=False, header=False)
```

```python
#p-value cutoff for setting α

p = 0.1

display(rra.loc[rra['padj'] < p].shape[0] / rra.shape[0])
```

```python
#α-RRA output

path = '/home/tyler/Documents/hDels/hDel-v2-α-RRA/'

hdel_rra_5_deseq = pd.read_csv(path + 'hDel-v2_DESeq2_α-RRA_out.txt', sep='\t')
```

```python
pattern = r"(.*)__(.*)"

def extract_pattern(df):
    df = df['group_id'].str.extract(pattern).rename(columns={0: 'hDel_ID', 1: 'hDel_window'}).join(df)
    return df

hdel_rra_5_deseq = extract_pattern(hdel_rra_5_deseq)
```
