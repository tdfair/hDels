# Import pseudobulk and metadata and perform DESeq2 likelihood-ratio test for each sgRNA pseudobulk (design = ~GEM well + sgRNA)

```r
for (sgRNA in all_sgRNAs$sg_ID) {

  #import pseudobulk counts and metadata
  pseudobulk_counts_fname <- paste0(sgRNA, "_pseudobulk.tsv")
  metadata_fname <- paste0(sgRNA, "_metadata.tsv")
  
  counts <- t(read.table(pseudobulk_counts_fname, header=TRUE, sep="\t", row.names=1))
  metadata <- read.table(metadata_fname, header=TRUE, sep="\t", row.names='SampleName')
  sgRNA_DESeq2 <- names(metadata)[names(metadata) != "batch"]

  #run DESeq2 likelihood ratio test
  DESeqDataSet <- DESeq2::DESeqDataSetFromMatrix(countData=counts, colData=metadata, 
                                                 design=formula(paste("~batch+", sgRNA_DESeq2)))
  lrt <- DESeq2::DESeq(DESeqDataSet, test="LRT", reduced=~batch, parallel=TRUE)
  #plotDispEsts(lrt, cex=0.6)
 
  #save results 
  results <- data.frame(DESeq2::results(lrt))
  results$sg_ID <- sgRNA_DESeq2
  
  DESeq2_fname <- paste0(f_out, sgRNA, "_DESeq2.csv")
  write.csv(results, DESeq2_fname, row.names=TRUE)
    
}
```
